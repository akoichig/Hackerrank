```python
from decimal import *


import sys

getcontext().prec=28
x,y = input().strip().split(' ')

# DEBUGGING:
N = 12 # 2 # 12
#x, y = N/2, N-1

x,y = [Decimal(x),Decimal(y)]
# your code goes here
#dist = {}
dist = []
sqdist = []


for i in [n for n in range(-N,N+1) if n!=0]:
    for j in [n for n in range(-N,N+1) if n!=0 and i**2+n**2 not in perfectSquares]:
        xi = Decimal(i)
        yj = Decimal(j)
        if (xi**2+yj**2).sqrt() % 1 != 0: # not an integer distance
            sqdist.append(i*i+j*j)
            dist.append((xi**2+yj**2).sqrt())
#dist[(xi, yj)]=(xi**2+yj**2).sqrt()
#print(min(dist.values()), max(dist.values())
#print(min(dist), max(dist))
#print(len(dist))
#print(len(set(dist)))

# Get distribution of values
ssd = sorted(sqdist)
ssdFreq = {}
for val in set(ssd):
    cnt =  sqdist.count(val)
    if cnt in ssdFreq:
        ssdFreq[cnt].append(val)
    else:
        ssdFreq[cnt] = [val]

for cnt in ssdFreq:
    print("Count: {}, No elements: {}\n     {}".format(cnt, len(ssdFreq[cnt]), ssdFreq[cnt]))
def fff():
    print("No. of pairs: {}".format(len(sqdist)))
    print("No of unique distances: {}".format(len(set(sqdist))))
    print(ssd)
    print(dist)
fff()    
f3 = dist[:3]
#print(sorted(f3, key=lambda x: x % 1))
for i in range(1, 13):
    for j in range(i, 13):
        if Decimal(i*i+j*j).sqrt() % 1 == 0:
            print(i, j, Decimal(i*i+j*j).sqrt())
```
